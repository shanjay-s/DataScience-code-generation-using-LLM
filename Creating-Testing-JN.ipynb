```python
import json

# Read the IPYNB file
with open("Testing-100rows.ipynb", mode="r", encoding="utf-8") as f:
    notebook_content = json.load(f)
```

```python
import pandas as pd
df = pd.read_csv("Prediction-mgc.csv")
```

```python
df["Acutal_answer"]=df["Acutal_answer"].str.replace("</code>","")
```

```python
df
```

```python
notebook_content
```

```python
for i in notebook_content:
    print(i)
```

```python
id_ind = 0
actual_ans_id = 1
predicted_ans_id = 2
result_empty_id = 3
for i in df.index:
    notebook_content["cells"][id_ind]["source"] = "#"+str(i)
    notebook_content["cells"][actual_ans_id]["source"] = df["Acutal_answer"][i]
    notebook_content["cells"][predicted_ans_id]["source"] = df["Predicted_answer"][i]
    notebook_content["cells"][result_empty_id]["source"] = ""
    id_ind += 4
    actual_ans_id+=4
    predicted_ans_id+=4
    result_empty_id+=4
```

```python
# Now you can manipulate the notebook_content (which is in JSON format)
# For example, you can modify cells, metadata, etc.

# To save changes back to a new IPYNB file:
with open("Testing-100rows.ipynb", mode="w", encoding="utf-8") as f:
    json.dump(notebook_content, f, indent=2)
```
